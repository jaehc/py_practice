# Find the limits and the basins of attractions
This is from the Calculus by Gibert Strang. <br> Find the limits and the basins of attractions for $F(x) = x - \tan { x }$.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def iteration(initial, n=10): 
    x = initial
    l = []
    for _ in range(n):
        x = x - np.tan(x)
        l.append(x)
    return l 

plt.figure(figsize=(20,5))

for c in np.linspace(np.pi*-3, np.pi*3, 100):
    end = iteration(c, 1000)[-1]
    plt.plot(c, end, 'o')

plt.tight_layout()
```

The limit grows as the initial point increases and there is a periodic pattern that occurs every pi. The limits are level between the interval, but diverse at the end of the period.
